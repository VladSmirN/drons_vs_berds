```python
import glob
import os
from tqdm import tqdm
from sahi.scripts.coco_evaluation import evaluate
```

```python
results_path = glob.glob('/home/vlad/projects/drons_vs_berds/prediction/*/results/*.*')  
```

```python
import json
dicts_results = [] 
for result_path in tqdm(results_path):
    basename = os.path.basename(result_path).split('.')[0]
    print(basename)
    evaluate(
        dataset_json_path='/home/vlad/datasets/drons_vs_berds/coco/dataset/drons_vs_berds_test.json' ,
        result_json_path=result_path,
        out_dir='./metrics'
    ) 
    with open('./metrics/eval.json') as f:
      data = json.load(f)

    data['name'] = basename
    dicts_results.append(data)
   
```
